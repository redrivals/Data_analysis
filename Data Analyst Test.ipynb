```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('level_progress.csv')
```

```python
df.head()
```

```python
df.describe
```

```python
def status_new(x):
    if x=='start':
        return 1
    if x == 'complete':
        return 2
    if x == 'fail':
        return 3
```

```python
df['Status_New'] = df['status'].apply(status_new)
```

```python
df
```

```python
df.loc[df.Status_New==3]
```

```python
fail = df.loc[df.Status_New==3]
```

```python
df.loc[df.Status_New==3].count()
```

```python
fail
```

```python
fail.plot(y ='level_number', x='Status_New', kind = 'scatter')	
```

```python
df.plot(y ='level_number', x='Status_New', kind = 'bar')
```
